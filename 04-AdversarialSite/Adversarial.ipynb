## Adversarial toolbox

Visit the site [Adversarial Toolbox](https://kennysong.github.io/adversarial.js/) to experiment with adversarial attacks and defenses. 

You can also run locally by starting a server in the 'adversarial.js/docs' directory. For example, you can use the following command in the terminal:

```
cd adversarial.js/docs
python3 -m http.server
```

Then, open a browser and go to `http://localhost:8000/`.
