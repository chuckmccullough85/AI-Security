## Adversarial toolbox

Visit the site [Adversarial Toolbox](https://kennysong.github.io/adversarial.js/) to experiment with adversarial attacks and defenses. 

The toolbox is a web-based tool that allows you to experiment with adversarial attacks and defenses on image classifiers. It is built using TensorFlow.js and React. The toolbox includes a variety of attacks and defenses, including FGSM, PGD, and adversarial training. You can experiment with different attacks and defenses on different models and datasets. The toolbox also includes a visualization tool that allows you to see the adversarial examples generated by the attacks.
